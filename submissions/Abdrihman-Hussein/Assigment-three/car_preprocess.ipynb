```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("raw_car_dataset.csv")
df.head()
```

```python
# initial info
df.isnull().sum()
```

```python
df ['Location'].value_counts()
```

```python
df['Location'] = df['Location'].replace({"Subrb": "Suburb", "??" : pd.NA})
df.head(5)
```

```python
df['Doors'] = df['Doors'].replace({ "" : pd.NA})
df['Odometer_km'] = df['Odometer_km'].replace({"" : pd.NA})
df.head(4)
```

```python
print ('clenin price null')
df['Price'] = df['Price'].replace(r"[\?,$]", "" , regex = True).astype('float')
df.head()
```

```python
df['Odometer_km'] = df['Odometer_km'].fillna(df['Odometer_km'].median())
df['Doors'] = df['Doors'].fillna(df['Doors'].mode()[0])
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
df
```

```python
# removing dublictes
df= df.drop_duplicates()
df.shape
```

```python
# cheaking info after cleaning
df.info()
```

```python
df ['Location'].value_counts()
```

```python
# ouliers
def iqr_fun(series, k=1.5):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

low_price, high_price = iqr_fun(df['Price'])
low_Odometer_km, high_Odometer_km = iqr_fun(df['Odometer_km'])

df['Price'] = df['Price'].clip(lower=low_price, upper=high_price)
df['Odometer_km'] = df['Odometer_km'].clip(lower=low_Odometer_km, upper=high_Odometer_km)
df
```

```python
# One-Hot Encoding
df = pd.get_dummies(
    df,
    columns=["Location"],
    dtype=int
)

print(df.columns)
```

```python
print(df.head())
```

```python
# Feature Engineering  

CURRENT_YEAR = 2026


df["Car_age"] = CURRENT_YEAR - df["Year"] 

df["Km_per_year"] = df["Odometer_km"] / df["Car_age"]
df["is_Urbun"] = df["Location_City"] + df["Location_Suburb"]
```

```python
# scaling 

dont_Scale = ["Price"]

numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns.to_list()

exclude_Features = [ c for c in df.columns if c.startswith("Location_")] + ["is_Urbun"] + ["is_Accident"]
num_features_to_scale = [c for c in numeric_cols if c not in dont_Scale and c not in exclude_Features]

scaler = StandardScaler()
df[num_features_to_scale] = scaler.fit_transform( df[num_features_to_scale])
```
